```python
# Задание №1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Загружаем данные
iris_data = load_iris()
df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
df['species'] = pd.Categorical.from_codes(iris_data.target, iris_data.target_names)

# =============================================================================
# ЗАДАНИЕ 1: Анализ выбросов для Iris virginica по признаку sepal width
# =============================================================================

# 1. СОЗДАЕМ ЯЩИКИ С УСАМИ (BOXPLOTS) ДЛЯ ВСЕХ ПРИЗНАКОВ ПО ВИДАМ
# -----------------------------------------------------------------------------
plt.figure(figsize=(12, 8))

# Создаем 4 подграфика (2 ряда по 2 графика) - по одному для каждого признака
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Анализ выбросов признаков Iris по видам', fontsize=16)

# Названия признаков для подписей
feature_names = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
feature_titles = ['Длина чашелистика', 'Ширина чашелистика', 'Длина лепестка', 'Ширина лепестка']

# Создаем boxplot для каждого признака
for i, (feature, title) in enumerate(zip(feature_names, feature_titles)):
    row, col = i // 2, i % 2  # Определяем позицию графика (2x2)

    # Группируем данные по видам для текущего признака
    data_to_plot = [df[df['species'] == species][feature] for species in df['species'].unique()]

    # Создаем boxplot
    bp = axes[row, col].boxplot(data_to_plot,
                                 labels=['setosa', 'versicolor', 'virginica'],
                                 patch_artist=True,  # Заливка цветом
                                 boxprops=dict(alpha=0.7),
                                 flierprops=dict(marker='o', markerfacecolor='red', markersize=8, alpha=0.5))  # Выбросы красным

    # Закрашиваем ящики разными цветами
    colors = ['lightblue', 'lightgreen', 'lightcoral']
    for patch, color in zip(bp['boxes'], colors):
        patch.set_facecolor(color)

    axes[row, col].set_title(f'{title}')
    axes[row, col].set_ylabel('см')
    axes[row, col].grid(True, linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()

# =============================================================================
# 2. АНАЛИЗ ВЫБРОСОВ ДЛЯ Iris virginica ПО ПРИЗНАКУ sepal width
# =============================================================================

# Выделяем данные только для вида virginica
virginica_data = df[df['species'] == 'virginica']['sepal width (cm)']

# Рассчитываем статистики для определения границ усов
Q1 = virginica_data.quantile(0.25)  # Первый квартиль (25-й перцентиль)
Q3 = virginica_data.quantile(0.75)  # Третий квартиль (75-й перцентиль)
IQR = Q3 - Q1  # Межквартильный размах

# Границы усов (обычно: Q1 - 1.5*IQR и Q3 + 1.5*IQR)
lower_whisker = Q1 - 1.5 * IQR
upper_whisker = Q3 + 1.5 * IQR

print("="*60)
print("АНАЛИЗ ВЫБРОСОВ ДЛЯ Iris virginica (sepal width)")
print("="*60)
print(f"Статистики для признака sepal width у вида virginica:")
print(f"  - Минимальное значение: {virginica_data.min():.3f}")
print(f"  - Максимальное значение: {virginica_data.max():.3f}")
print(f"  - Среднее значение: {virginica_data.mean():.3f}")
print(f"  - Медиана (Q2): {virginica_data.median():.3f}")
print(f"  - Первый квартиль (Q1): {Q1:.3f}")
print(f"  - Третий квартиль (Q3): {Q3:.3f}")
print(f"  - Межквартильный размах (IQR): {IQR:.3f}")
print(f"\nГраницы усов:")
print(f"  - Нижняя граница (Q1 - 1.5*IQR): {lower_whisker:.3f}")
print(f"  - Верхняя граница (Q3 + 1.5*IQR): {upper_whisker:.3f}")

# Находим выбросы
outliers = virginica_data[(virginica_data < lower_whisker) | (virginica_data > upper_whisker)]
outliers_count = len(outliers)
total_count = len(virginica_data)
outliers_percent = (outliers_count / total_count) * 100

print(f"\nРЕЗУЛЬТАТЫ АНАЛИЗА ВЫБРОСОВ:")
print(f"  - Количество наблюдений virginica: {total_count}")
print(f"  - Количество выбросов: {outliers_count}")
print(f"  - Процент выбросов: {outliers_percent:.2f}%")

if outliers_count > 0:
    print(f"\nЗначения выбросов:")
    for idx, value in outliers.items():
        print(f"    * Строка {idx}: {value:.3f} см")
else:
    print(f"\nВыбросы не обнаружены.")

# Визуализируем распределение с выделением выбросов
plt.figure(figsize=(10, 6))

# Гистограмма распределения
plt.hist(virginica_data, bins=10, alpha=0.7, color='lightcoral', edgecolor='black', label='Все наблюдения')

# Отмечаем выбросы на гистограмме
if outliers_count > 0:
    plt.hist(outliers, bins=5, alpha=0.8, color='red', edgecolor='black', label='Выбросы')

# Добавляем вертикальные линии для границ
plt.axvline(x=lower_whisker, color='blue', linestyle='--', linewidth=2, label=f'Нижняя граница ({lower_whisker:.3f})')
plt.axvline(x=upper_whisker, color='blue', linestyle='--', linewidth=2, label=f'Верхняя граница ({upper_whisker:.3f})')
plt.axvline(x=Q1, color='green', linestyle=':', linewidth=1.5, alpha=0.7, label=f'Q1 ({Q1:.3f})')
plt.axvline(x=Q3, color='green', linestyle=':', linewidth=1.5, alpha=0.7, label=f'Q3 ({Q3:.3f})')
plt.axvline(x=virginica_data.median(), color='purple', linestyle='-', linewidth=2, label=f'Медиана ({virginica_data.median():.3f})')

plt.title('Распределение ширины чашелистика (sepal width) у Iris virginica', fontsize=14)
plt.xlabel('Ширина чашелистика (см)')
plt.ylabel('Частота')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# =============================================================================
#Сравнение с другими видами
# =============================================================================

print("\n" + "="*60)
print("СРАВНЕНИЕ С ДРУГИМИ ВИДАМИ")
print("="*60)

for species in df['species'].unique():
    species_data = df[df['species'] == species]['sepal width (cm)']
    Q1_s = species_data.quantile(0.25)
    Q3_s = species_data.quantile(0.75)
    IQR_s = Q3_s - Q1_s
    lower_s = Q1_s - 1.5 * IQR_s
    upper_s = Q3_s + 1.5 * IQR_s
    outliers_s = species_data[(species_data < lower_s) | (species_data > upper_s)]

    print(f"\n{species.capitalize()}:")
    print(f"  - Диапазон: [{species_data.min():.3f}, {species_data.max():.3f}]")
    print(f"  - Q1, Q3: [{Q1_s:.3f}, {Q3_s:.3f}]")
    print(f"  - IQR: {IQR_s:.3f}")
    print(f"  - Границы усов: [{lower_s:.3f}, {upper_s:.3f}]")
    print(f"  - Выбросы: {len(outliers_s)} ({len(outliers_s)/len(species_data)*100:.1f}%)")
```

```python
# =============================================================================
# ЗАДАНИЕ 2: Расчет коэффициента вариации для всех признаков
# =============================================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Загружаем данные
iris_data = load_iris()
df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
df['species'] = pd.Categorical.from_codes(iris_data.target, iris_data.target_names)



# Список числовых признаков (исключаем species)
feature_columns = iris_data.feature_names

print("="*70)
print("КОЭФФИЦИЕНТ ВАРИАЦИИ ДЛЯ ПРИЗНАКОВ IRIS")
print("="*70)

# Создаем список для хранения результатов
cv_results = []

# Рассчитываем коэффициент вариации для каждого признака
for feature in feature_columns:
    # Извлекаем данные по признаку
    feature_data = df[feature]

    # Рассчитываем среднее и стандартное отклонение
    mean_val = feature_data.mean()
    std_val = feature_data.std()

    # Коэффициент вариации = (std / mean) * 100%
    cv = (std_val / mean_val) * 100

    cv_results.append({
        'Признак': feature,
        'Среднее': mean_val,
        'Ст. отклонение': std_val,
        'CV (%)': cv
    })

    print(f"\n{feature}:")
    print(f"  Среднее значение: {mean_val:.3f}")
    print(f"  Стандартное отклонение: {std_val:.3f}")
    print(f"  Коэффициент вариации: {cv:.2f}%")

# Создаем DataFrame с результатами для удобного вывода
cv_df = pd.DataFrame(cv_results)
cv_df = cv_df.sort_values('CV (%)', ascending=False)

print("\n" + "="*70)
print("ПРИЗНАКИ, ОТСОРТИРОВАННЫЕ ПО КОЭФФИЦИЕНТУ ВАРИАЦИИ:")
print("="*70)
print(cv_df.to_string(index=False))

# =============================================================================
# ВИЗУАЛИЗАЦИЯ КОЭФФИЦИЕНТОВ ВАРИАЦИИ
# =============================================================================

plt.figure(figsize=(10, 6))

# Создаем столбчатую диаграмму
bars = plt.bar(cv_df['Признак'], cv_df['CV (%)'], color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'])

# Добавляем значения над столбцами
for bar, cv in zip(bars, cv_df['CV (%)']):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.5,
             f'{cv:.1f}%', ha='center', va='bottom', fontsize=11)

plt.title('Коэффициент вариации признаков датасета Iris', fontsize=15)
plt.ylabel('Коэффициент вариации (%)', fontsize=12)
plt.xticks(rotation=15, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()

# =============================================================================
# АНАЛИЗ ИЗМЕНЧИВОСТИ ПО ВИДАМ
# =============================================================================

print("\n" + "="*70)
print("АНАЛИЗ ИЗМЕНЧИВОСТИ В РАЗРЕЗЕ ВИДОВ:")
print("="*70)

# Рассчитываем CV для каждого вида отдельно
for species in df['species'].unique():
    print(f"\n{species.capitalize()}:")
    species_df = df[df['species'] == species]

    for feature in feature_columns:
        feature_data = species_df[feature]
        cv_species = (feature_data.std() / feature_data.mean()) * 100
        print(f"  {feature}: CV = {cv_species:.2f}%")

# =============================================================================
# ВИЗУАЛИЗАЦИЯ РАСПРЕДЕЛЕНИЙ ДЛЯ СРАВНЕНИЯ
# =============================================================================

# Создаем график распределений всех признаков
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Распределение признаков Iris (нормализованные гистограммы)', fontsize=16)

feature_titles = ['Длина чашелистика', 'Ширина чашелистика', 'Длина лепестка', 'Ширина лепестка']
colors = ['red', 'green', 'blue']

for i, (feature, title) in enumerate(zip(feature_columns, feature_titles)):
    row, col = i // 2, i % 2

    # Нормализуем данные для лучшего сравнения (делим на среднее)
    for species, color in zip(df['species'].unique(), colors):
        species_data = df[df['species'] == species][feature]
        # Нормализуем относительно среднего по виду для наглядности
        normalized_data = species_data / species_data.mean()

        axes[row, col].hist(normalized_data, bins=8, alpha=0.5,
                           color=color, label=species.capitalize(), density=True)

    axes[row, col].set_title(title)
    axes[row, col].set_xlabel('Значение / Среднее')
    axes[row, col].set_ylabel('Плотность')
    axes[row, col].legend()
    axes[row, col].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# =============================================================================
# ВАЖНОСТЬ КОЭФФИЦИЕНТА ВАРИАЦИИ ДЛЯ НОРМАЛИЗАЦИИ
# =============================================================================

print("\n" + "="*70)
print("ПОЧЕМУ ЭТО ВАЖНО ДЛЯ НОРМАЛИЗАЦИИ ДАННЫХ?")
print("="*70)

# Находим признак с максимальным CV
max_cv_feature = cv_df.iloc[0]['Признак']
max_cv_value = cv_df.iloc[0]['CV (%)']

print(f"\nНаиболее изменчивый признак: {max_cv_feature}")
print(f"Коэффициент вариации: {max_cv_value:.2f}%")

print("""
Почему анализ коэффициента вариации важен для нормализации данных:

1. МАСШТАБИРОВАНИЕ ПРИЗНАКОВ:
   - Признаки с высоким CV имеют большой разброс относительно среднего
   - При использовании алгоритмов, основанных на расстояниях (k-means, PCA, SVM),
     признаки с большим разбросом будут доминировать над признаками с малым разбросом
   - Нормализация приводит все признаки к сопоставимому масштабу

2. ИНТЕРПРЕТАЦИЯ РЕЗУЛЬТАТОВ:
   - Высокий CV может указывать на наличие подгрупп в данных (как в Iris)
   - В нашем случае, признаки лепестка более изменчивы, т.к. сильно различаются между видами
   - Это естественно, ведь именно по лепесткам виды ирисов легче всего различить
""")

# Демонстрация влияния нормализации
print("\n" + "="*70)
print("ДЕМОНСТРАЦИЯ ВЛИЯНИЯ НОРМАЛИЗАЦИИ:")
print("="*70)

from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Выбираем два самых контрастных признака
feature1 = 'petal length (cm)'  # Самый изменчивый
feature2 = 'sepal width (cm)'    # Самый стабильный

# Создаем копию данных для демонстрации
demo_data = df[[feature1, feature2]].copy()

print(f"\nДо нормализации:")
print(f"  {feature1}: mean={demo_data[feature1].mean():.3f}, std={demo_data[feature1].std():.3f}")
print(f"  {feature2}: mean={demo_data[feature2].mean():.3f}, std={demo_data[feature2].std():.3f}")

# Стандартизация (Z-score)
scaler_std = StandardScaler()
demo_std = scaler_std.fit_transform(demo_data)
demo_std_df = pd.DataFrame(demo_std, columns=[f'{f}_std' for f in [feature1, feature2]])

print(f"\nПосле стандартизации (Z-score):")
print(f"  {feature1}: mean={demo_std_df[f'{feature1}_std'].mean():.3f}, std={demo_std_df[f'{feature1}_std'].std():.3f}")
print(f"  {feature2}: mean={demo_std_df[f'{feature2}_std'].mean():.3f}, std={demo_std_df[f'{feature2}_std'].std():.3f}")

# Min-Max нормализация
scaler_mm = MinMaxScaler()
demo_mm = scaler_mm.fit_transform(demo_data)
demo_mm_df = pd.DataFrame(demo_mm, columns=[f'{f}_mm' for f in [feature1, feature2]])

print(f"\nПосле Min-Max нормализации:")
print(f"  {feature1}: min={demo_mm_df[f'{feature1}_mm'].min():.3f}, max={demo_mm_df[f'{feature1}_mm'].max():.3f}")
print(f"  {feature2}: min={demo_mm_df[f'{feature2}_mm'].min():.3f}, max={demo_mm_df[f'{feature2}_mm'].max():.3f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Загружаем данные
iris_data = load_iris()
df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
df['species'] = pd.Categorical.from_codes(iris_data.target, iris_data.target_names)

# =============================================================================
# ЗАДАНИЕ 3: Создание признака "соотношение длины и ширины лепестка"
# =============================================================================

print("="*70)
print("ЗАДАНИЕ 3: СОЗДАНИЕ ПРИЗНАКА 'СООТНОШЕНИЕ ДЛИНЫ И ШИРИНЫ ЛЕПЕСТКА'")
print("="*70)

# Создаем новый признак - отношение длины лепестка к его ширине
# Добавляем маленькую константу, чтобы избежать деления на ноль (хотя в Iris ширина лепестка > 0)
df['petal_length_width_ratio'] = df['petal length (cm)'] / df['petal width (cm)']

print("\nПервые 10 строк с новым признаком:")
print(df[['species', 'petal length (cm)', 'petal width (cm)', 'petal_length_width_ratio']].head(10))

# =============================================================================
# СТАТИСТИКА НОВОГО ПРИЗНАКА ПО ВИДАМ
# =============================================================================

print("\n" + "="*70)
print("СТАТИСТИКА ПРИЗНАКА 'СООТНОШЕНИЕ ДЛИНЫ/ШИРИНЫ ЛЕПЕСТКА' ПО ВИДАМ:")
print("="*70)

# Группируем по видам и считаем статистики
ratio_stats = df.groupby('species')['petal_length_width_ratio'].agg(['count', 'mean', 'std', 'min', 'max'])
print("\n", ratio_stats.round(3))

# =============================================================================
# ГИСТОГРАММА НОВОГО ПРИЗНАКА ПО ВИДАМ
# =============================================================================

plt.figure(figsize=(12, 8))

# Создаем гистограмму с разбивкой по видам
colors = ['red', 'green', 'blue']
species_list = df['species'].unique()

# Используем плотность (density=True) для лучшего сравнения, т.к. выборки одинакового размера
for species, color in zip(species_list, colors):
    species_data = df[df['species'] == species]['petal_length_width_ratio']
    plt.hist(species_data, bins=15, alpha=0.6, color=color,
             label=species.capitalize(), density=True, edgecolor='black')

plt.title('Распределение соотношения длины и ширины лепестка по видам ирисов', fontsize=15)
plt.xlabel('Соотношение (длина / ширина лепестка)', fontsize=12)
plt.ylabel('Плотность вероятности', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3, linestyle='--')

# Добавляем вертикальные линии для средних значений
for species, color in zip(species_list, colors):
    mean_val = df[df['species'] == species]['petal_length_width_ratio'].mean()
    plt.axvline(x=mean_val, color=color, linestyle='--', linewidth=2, alpha=0.8)
    plt.text(mean_val + 0.1, 0.02, f'{species.capitalize()}: {mean_val:.2f}',
             rotation=90, fontsize=10, color=color)

plt.tight_layout()
plt.show()

# =============================================================================
# СРАВНЕНИЕ РАЗДЕЛИМОСТИ КЛАССОВ
# =============================================================================

print("\n" + "="*70)
print("СРАВНЕНИЕ РАЗДЕЛИМОСТИ КЛАССОВ:")
print("="*70)

# Создаем матрицу 2x2 для сравнения исходных признаков и нового признака
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
fig.suptitle('Сравнение разделимости классов: исходные признаки vs новый признак', fontsize=16)

# 1. Длина лепестка (исходный признак)
ax1 = axes[0, 0]
for species, color in zip(species_list, colors):
    species_data = df[df['species'] == species]['petal length (cm)']
    ax1.hist(species_data, bins=15, alpha=0.6, color=color,
             label=species.capitalize(), density=True, edgecolor='black')
ax1.set_title('Длина лепестка (исходный признак)')
ax1.set_xlabel('Длина лепестка (см)')
ax1.set_ylabel('Плотность')
ax1.legend()
ax1.grid(True, alpha=0.3)

# 2. Ширина лепестка (исходный признак)
ax2 = axes[0, 1]
for species, color in zip(species_list, colors):
    species_data = df[df['species'] == species]['petal width (cm)']
    ax2.hist(species_data, bins=15, alpha=0.6, color=color,
             label=species.capitalize(), density=True, edgecolor='black')
ax2.set_title('Ширина лепестка (исходный признак)')
ax2.set_xlabel('Ширина лепестка (см)')
ax2.set_ylabel('Плотность')
ax2.legend()
ax2.grid(True, alpha=0.3)

# 3. Соотношение (новый признак)
ax3 = axes[1, 0]
for species, color in zip(species_list, colors):
    species_data = df[df['species'] == species]['petal_length_width_ratio']
    ax3.hist(species_data, bins=15, alpha=0.6, color=color,
             label=species.capitalize(), density=True, edgecolor='black')
ax3.set_title('Соотношение длина/ширина лепестка (новый признак)')
ax3.set_xlabel('Соотношение')
ax3.set_ylabel('Плотность')
ax3.legend()
ax3.grid(True, alpha=0.3)

# 4. Диаграмма рассеяния для визуализации разделимости
ax4 = axes[1, 1]
for species, color in zip(species_list, colors):
    species_data = df[df['species'] == species]
    ax4.scatter(species_data['petal length (cm)'], species_data['petal width (cm)'],
                c=color, label=species.capitalize(), alpha=0.7, edgecolors='black', linewidth=0.5)

# Добавляем линии константного соотношения
x_range = np.linspace(0, 7, 100)
for ratio in [2, 3, 4, 5]:
    ax4.plot(x_range, x_range/ratio, 'k--', alpha=0.3, linewidth=1)
    ax4.text(6, 6/ratio + 0.1, f'ratio={ratio}', fontsize=8, alpha=0.5)

ax4.set_title('Соотношение длина/ширина (диаграмма рассеяния)')
ax4.set_xlabel('Длина лепестка (см)')
ax4.set_ylabel('Ширина лепестка (см)')
ax4.legend()
ax4.grid(True, alpha=0.3)
ax4.set_xlim(0, 7.5)
ax4.set_ylim(0, 3)

plt.tight_layout()
plt.show()

# =============================================================================
# КОЛИЧЕСТВЕННАЯ ОЦЕНКА РАЗДЕЛИМОСТИ
# =============================================================================

print("\n" + "="*70)
print("КОЛИЧЕСТВЕННАЯ ОЦЕНКА РАЗДЕЛИМОСТИ:")
print("="*70)

def calculate_separation_metrics(data, feature):
    """
    Рассчитывает метрики разделимости для признака
    """
    setosa = data[data['species'] == 'setosa'][feature]
    versicolor = data[data['species'] == 'versicolor'][feature]
    virginica = data[data['species'] == 'virginica'][feature]

    # Расстояние между средними (нормализованное на стандартное отклонение)
    mean_dist_sv = abs(setosa.mean() - versicolor.mean()) / ((setosa.std() + versicolor.std()) / 2)
    mean_dist_svir = abs(setosa.mean() - virginica.mean()) / ((setosa.std() + virginica.std()) / 2)
    mean_dist_vvir = abs(versicolor.mean() - virginica.mean()) / ((versicolor.std() + virginica.std()) / 2)

    # Перекрытие распределений (процент перекрытия оцениваем приблизительно)
    # Чем меньше перекрытие, тем лучше разделение

    return {
        'Среднее setosa': setosa.mean(),
        'Среднее versicolor': versicolor.mean(),
        'Среднее virginica': virginica.mean(),
        'Std setosa': setosa.std(),
        'Std versicolor': versicolor.std(),
        'Std virginica': virginica.std(),
        'Расстояние setosa-versicolor': mean_dist_sv,
        'Расстояние setosa-virginica': mean_dist_svir,
        'Расстояние versicolor-virginica': mean_dist_vvir,
        'Среднее расстояние': (mean_dist_sv + mean_dist_svir + mean_dist_vvir) / 3
    }

# Сравниваем признаки
features_to_compare = ['petal length (cm)', 'petal width (cm)', 'petal_length_width_ratio']

print("\nСравнительные метрики разделимости:")
for feature in features_to_compare:
    metrics = calculate_separation_metrics(df, feature)
    print(f"\n{feature}:")
    print(f"  Среднее расстояние между классами: {metrics['Среднее расстояние']:.3f}")
    print(f"  setosa vs versicolor: {metrics['Расстояние setosa-versicolor']:.3f}")
    print(f"  setosa vs virginica: {metrics['Расстояние setosa-virginica']:.3f}")
    print(f"  versicolor vs virginica: {metrics['Расстояние versicolor-virginica']:.3f}")

# =============================================================================
# ВИЗУАЛИЗАЦИЯ С ПОМОЩЬЯ BOXPLOT
# =============================================================================

plt.figure(figsize=(12, 6))

# Создаем boxplot для сравнения
data_to_plot = [
    df[df['species'] == 'setosa']['petal_length_width_ratio'],
    df[df['species'] == 'versicolor']['petal_length_width_ratio'],
    df[df['species'] == 'virginica']['petal_length_width_ratio']
]

bp = plt.boxplot(data_to_plot, labels=['setosa', 'versicolor', 'virginica'],
                 patch_artist=True, widths=0.6)

# Раскрашиваем ящики
colors = ['red', 'green', 'blue']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.6)

plt.title('Сравнение соотношения длина/ширина лепестка по видам (boxplot)', fontsize=14)
plt.xlabel('Вид ириса', fontsize=12)
plt.ylabel('Соотношение (длина / ширина лепестка)', fontsize=12)
plt.grid(True, axis='y', alpha=0.3)

# Добавляем точки всех наблюдений (strip plot)
for i, species in enumerate(['setosa', 'versicolor', 'virginica'], 1):
    y = df[df['species'] == species]['petal_length_width_ratio']
    x = np.random.normal(i, 0.05, size=len(y))  # Добавляем небольшой шум для разброса точек
    plt.scatter(x, y, alpha=0.5, color='black', s=30)

plt.tight_layout()
plt.show()

# =============================================================================
# ВЫВОДЫ
# =============================================================================

print("\n" + "="*70)
print("ВЫВОДЫ:")
print("="*70)

print("""
РАЗДЕЛИМОСТЬ КЛАССОВ:
   - Исходные признаки (длина и ширина лепестка) уже хорошо разделяют Iris setosa
   - Однако между versicolor и virginica есть некоторое перекрытие

НОВЫЙ ПРИЗНАК (соотношение длина/ширина):
   - Setosa имеет наименьшее соотношение (около 2.5), т.е. лепестки более округлые
   - Versicolor имеет среднее соотношение (около 3.5-4.0)
   - Virginica имеет наибольшее соотношение (около 4.5-5.5), т.е. лепестки более вытянутые

УЛУЧШЕНИЕ РАЗДЕЛИМОСТИ:
   - Между setosa и другими видами: разделение остается отличным
   - Между versicolor и virginica: перекрытие УМЕНЬШАЕТСЯ по сравнению с исходными признаками
   - Соотношение подчеркивает форму лепестка, которая является видоспецифичной

ПРЕИМУЩЕСТВА НОВОГО ПРИЗНАКА:
   - Инвариантен к масштабу (если бы цветы были сфотографированы с разным увеличением)
   - Отражает биологически значимую характеристику - форму лепестка
   - Улучшает разделение трудноразличимых классов versicolor и virginica
""")

# Демонстрация улучшения разделения versicolor и virginica
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Только versicolor и virginica для исходных признаков
versicolor_data = df[df['species'] == 'versicolor']
virginica_data = df[df['species'] == 'virginica']

ax1.scatter(versicolor_data['petal length (cm)'], versicolor_data['petal width (cm)'],
            c='green', label='versicolor', alpha=0.7, edgecolors='black', s=50)
ax1.scatter(virginica_data['petal length (cm)'], virginica_data['petal width (cm)'],
            c='blue', label='virginica', alpha=0.7, edgecolors='black', s=50)
ax1.set_xlabel('Длина лепестка (см)')
ax1.set_ylabel('Ширина лепестка (см)')
ax1.set_title('Исходные признаки (versicolor vs virginica)')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Распределение нового признака для versicolor и virginica
ax2.hist(versicolor_data['petal_length_width_ratio'], bins=12, alpha=0.7,
         color='green', label='versicolor', edgecolor='black', density=True)
ax2.hist(virginica_data['petal_length_width_ratio'], bins=12, alpha=0.7,
         color='blue', label='virginica', edgecolor='black', density=True)
ax2.set_xlabel('Соотношение длина/ширина лепестка')
ax2.set_ylabel('Плотность')
ax2.set_title('Новый признак (versicolor vs virginica)')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

# Загружаем датасет Wine
wine_data = load_wine()
df = pd.DataFrame(data=wine_data.data, columns=wine_data.feature_names)

print("="*80)
print("ДАТАСЕТ WINE: информация")
print("="*80)
print(f"Размерность данных: {df.shape}")
print(f"Признаки: {', '.join(wine_data.feature_names)}")

# =============================================================================
# ЗАДАНИЕ 1: ВЫЯВЛЕНИЕ МУЛЬТИКОЛЛИНЕАРНОСТИ (|r| > 0.7)
# =============================================================================

print("\n" + "="*80)
print("ЗАДАНИЕ 1: ВЫЯВЛЕНИЕ МУЛЬТИКОЛЛИНЕАРНОСТИ (|r| > 0.7)")
print("="*80)

# Строим матрицу корреляций
correlation_matrix = df.corr()

# Находим все пары с |r| > 0.7
print("\nПоиск пар признаков с |r| > 0.7:")

# Получаем верхний треугольник матрицы корреляций
corr_values = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))

# Преобразуем в длинный формат
corr_long = corr_values.stack().reset_index()
corr_long.columns = ['Признак_1', 'Признак_2', 'Корреляция']

# Фильтруем по |r| > 0.7
high_corr_df = corr_long[abs(corr_long['Корреляция']) > 0.7].copy()
high_corr_df = high_corr_df.sort_values('Корреляция', ascending=False)

print(f"\nНайдено {len(high_corr_df)} пар с |r| > 0.7")
print("\nТаблица сильно коррелирующих признаков:")
print(high_corr_df.to_string(index=False))

# Подсчитываем участие признаков
print("\n" + "="*80)
print("АНАЛИЗ УЧАСТИЯ ПРИЗНАКОВ В СИЛЬНЫХ КОРРЕЛЯЦИЯХ:")
print("="*80)

# Создаем серию со всеми признаками из обеих колонок
all_features = pd.concat([
    high_corr_df['Признак_1'],
    high_corr_df['Признак_2']
])

# Подсчитываем частоту каждого признака
correlation_count = all_features.value_counts().reset_index()
correlation_count.columns = ['Признак', 'Количество_сильных_связей']

print("\nКоличество сильных корреляций для каждого признака:")
print(correlation_count.to_string(index=False))

# Находим признак с максимальным числом связей
max_feature = correlation_count.iloc[0]['Признак']
max_count = correlation_count.iloc[0]['Количество_сильных_связей']

print(f"\n👉 Признак с наибольшим числом сильных корреляций: '{max_feature}' ({max_count} связей)")

# =============================================================================
# ВИЗУАЛИЗАЦИЯ ТОП-2 САМЫХ СИЛЬНЫХ КОРРЕЛЯЦИЙ
# =============================================================================

print("\n" + "="*80)
print("ВИЗУАЛИЗАЦИЯ ТОП-2 САМЫХ СИЛЬНЫХ КОРРЕЛЯЦИЙ:")
print("="*80)

# Берем топ-2 самых сильных корреляций
top_correlations = high_corr_df.head(2).copy()
top_correlations['R²'] = top_correlations['Корреляция'] ** 2
top_correlations['Объясненная_вариация_%'] = (top_correlations['R²'] * 100).round(1)

print("\nТоп-2 самых сильных корреляций:")
print(top_correlations[['Признак_1', 'Признак_2', 'Корреляция', 'R²', 'Объясненная_вариация_%']].to_string(index=False))

# Создаем два графика рядом
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Две самые сильные корреляции в датасете Wine', fontsize=16)

# Преобразуем в список пар для визуализации
top_pairs = top_correlations[['Признак_1', 'Признак_2']].values

# Заполняем графики
for idx, (feat1, feat2) in enumerate(top_pairs):
    ax = axes[idx]

    # Создаем scatter plot
    ax.scatter(df[feat1], df[feat2], alpha=0.7, s=60, c='steelblue', edgecolors='black', linewidth=0.5)
    ax.set_xlabel(feat1, fontsize=11)
    ax.set_ylabel(feat2, fontsize=11)

    # Добавляем линию регрессии
    z = np.polyfit(df[feat1], df[feat2], 1)
    p = np.poly1d(z)
    x_line = np.linspace(df[feat1].min(), df[feat1].max(), 100)
    ax.plot(x_line, p(x_line), "r-", alpha=0.8, linewidth=2.5)

    # Добавляем информацию о корреляции
    corr_value = correlation_matrix.loc[feat1, feat2]
    ax.set_title(f'{feat1} vs {feat2}\nr = {corr_value:.3f}',
                fontsize=12, fontweight='bold')
    ax.grid(True, alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()

# =============================================================================
# МАТРИЦА КОРРЕЛЯЦИЙ С ВЫДЕЛЕНИЕМ СИЛЬНЫХ СВЯЗЕЙ
# =============================================================================

# Создаем фигуру с двумя подграфиками
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

# Первый график: полная матрица корреляций
im1 = ax1.imshow(correlation_matrix, cmap='RdBu_r', vmin=-1, vmax=1, aspect='auto')
plt.colorbar(im1, ax=ax1, label='Корреляция', fraction=0.046)
ax1.set_xticks(range(len(correlation_matrix.columns)))
ax1.set_yticks(range(len(correlation_matrix.columns)))
ax1.set_xticklabels(correlation_matrix.columns, rotation=90, fontsize=8)
ax1.set_yticklabels(correlation_matrix.columns, fontsize=8)
ax1.set_title('Полная матрица корреляций', fontsize=14)

# Добавляем значения
for i in range(len(correlation_matrix.columns)):
    for j in range(len(correlation_matrix.columns)):
        if i != j:
            ax1.text(j, i, f'{correlation_matrix.iloc[i, j]:.2f}',
                    ha='center', va='center', fontsize=6)

# Второй график: только сильные корреляции
corr_with_highlight = correlation_matrix.copy()
corr_with_highlight[abs(corr_with_highlight) <= 0.7] = np.nan

im2 = ax2.imshow(corr_with_highlight, cmap='RdBu_r', vmin=-1, vmax=1, aspect='auto')
plt.colorbar(im2, ax=ax2, label='|r| > 0.7', fraction=0.046)
ax2.set_xticks(range(len(corr_with_highlight.columns)))
ax2.set_yticks(range(len(corr_with_highlight.columns)))
ax2.set_xticklabels(corr_with_highlight.columns, rotation=90, fontsize=8)
ax2.set_yticklabels(corr_with_highlight.columns, fontsize=8)
ax2.set_title('Только сильные корреляции (|r| > 0.7)', fontsize=14)

# Добавляем значения для сильных корреляций
for i in range(len(corr_with_highlight.columns)):
    for j in range(len(corr_with_highlight.columns)):
        if i != j and not np.isnan(corr_with_highlight.iloc[i, j]):
            text_color = 'white' if abs(corr_with_highlight.iloc[i, j]) > 0.5 else 'black'
            ax2.text(j, i, f'{corr_with_highlight.iloc[i, j]:.2f}',
                    ha='center', va='center', fontsize=8, fontweight='bold',
                    color=text_color)

plt.tight_layout()
plt.show()

# =============================================================================
# ВИЗУАЛИЗАЦИЯ СВЯЗЕЙ ПРИЗНАКА-ЛИДЕРА
# =============================================================================

print("\n" + "="*80)
print(f"ВИЗУАЛИЗАЦИЯ СВЯЗЕЙ ПРИЗНАКА-ЛИДЕРА: '{max_feature}'")
print("="*80)

# Получаем все корреляции признака-лидера с другими
leader_corrs = correlation_matrix[max_feature].drop(max_feature).reset_index()
leader_corrs.columns = ['Признак', 'Корреляция']
leader_corrs = leader_corrs.sort_values('Корреляция', ascending=False)

# Находим признаки, с которыми у лидера сильная корреляция
strong_leader_corrs = leader_corrs[abs(leader_corrs['Корреляция']) > 0.7]

print(f"\nПризнаки с |r| > 0.7 с '{max_feature}':")
if len(strong_leader_corrs) > 0:
    print(strong_leader_corrs.to_string(index=False))
else:
    print("Нет таких признаков")

# Создаем визуализацию для признака-лидера
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle(f'Анализ признака-лидера: "{max_feature}"', fontsize=16)

# График 1: Все корреляции признака-лидера (столбчатая диаграмма)
ax1 = axes[0]
colors = ['red' if abs(x) > 0.7 else 'skyblue' for x in leader_corrs['Корреляция']]
leader_corrs.plot(x='Признак', y='Корреляция', kind='bar', ax=ax1, color=colors, legend=False)
ax1.axhline(y=0.7, color='red', linestyle='--', alpha=0.5, label='|r| = 0.7')
ax1.axhline(y=-0.7, color='red', linestyle='--', alpha=0.5)
ax1.set_ylabel('Корреляция')
ax1.set_title(f'Корреляции признака "{max_feature}" с другими', fontsize=12)
ax1.tick_params(axis='x', rotation=90)
ax1.legend()
ax1.grid(True, alpha=0.3, axis='y')

# График 2: Scatter для самого сильного коррелята
ax2 = axes[1]
if len(leader_corrs) > 0:
    top_feature = leader_corrs.iloc[0]['Признак']
    corr_value = leader_corrs.iloc[0]['Корреляция']

    ax2.scatter(df[top_feature], df[max_feature], alpha=0.7, s=50,
                c='green', edgecolors='black', linewidth=0.5)

    # Линия регрессии
    z = np.polyfit(df[top_feature], df[max_feature], 1)
    p = np.poly1d(z)
    x_line = np.linspace(df[top_feature].min(), df[top_feature].max(), 100)
    ax2.plot(x_line, p(x_line), "r-", linewidth=2)

    ax2.set_xlabel(top_feature, fontsize=11)
    ax2.set_ylabel(max_feature, fontsize=11)
    ax2.set_title(f'{top_feature} vs {max_feature}\nr = {corr_value:.3f}', fontsize=12)
    ax2.grid(True, alpha=0.3)

# График 3: Scatter для второго по силе коррелята
ax3 = axes[2]
if len(leader_corrs) > 1:
    second_feature = leader_corrs.iloc[1]['Признак']
    corr_value = leader_corrs.iloc[1]['Корреляция']

    ax3.scatter(df[second_feature], df[max_feature], alpha=0.7, s=50,
                c='orange', edgecolors='black', linewidth=0.5)

    # Линия регрессии
    z = np.polyfit(df[second_feature], df[max_feature], 1)
    p = np.poly1d(z)
    x_line = np.linspace(df[second_feature].min(), df[second_feature].max(), 100)
    ax3.plot(x_line, p(x_line), "r-", linewidth=2)

    ax3.set_xlabel(second_feature, fontsize=11)
    ax3.set_ylabel(max_feature, fontsize=11)
    ax3.set_title(f'{second_feature} vs {max_feature}\nr = {corr_value:.3f}', fontsize=12)
    ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# =============================================================================
# СТАТИСТИЧЕСКИЙ ОТЧЕТ
# =============================================================================

print("\n" + "="*80)
print("СТАТИСТИЧЕСКИЙ ОТЧЕТ ПО МУЛЬТИКОЛЛИНЕАРНОСТИ")
print("="*80)

# Дополнительная статистика
total_pairs = len(df.columns) * (len(df.columns) - 1) // 2
print(f"\nОбщая статистика:")
print(f"  - Всего признаков: {len(df.columns)}")
print(f"  - Всего возможных пар: {total_pairs}")
print(f"  - Пар с |r| > 0.7: {len(high_corr_df)}")
print(f"  - Доля сильных связей: {len(high_corr_df) / total_pairs * 100:.1f}%")

# Анализ распределения корреляций
corr_upper = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))
all_correlations = corr_upper.stack().values

print(f"\nРаспределение корреляций:")
print(f"  - Минимальная: {np.nanmin(all_correlations):.3f}")
print(f"  - Максимальная: {np.nanmax(all_correlations):.3f}")
print(f"  - Средняя: {np.nanmean(all_correlations):.3f}")
print(f"  - Медианная: {np.nanmedian(all_correlations):.3f}")

# Распределение по интервалам
bins = [-1, -0.7, -0.3, 0.3, 0.7, 1]
labels = ['сильные отр.', 'средние отр.', 'слабые', 'средние пол.', 'сильные пол.']
corr_dist = pd.cut(all_correlations, bins=bins, labels=labels).value_counts().sort_index()

print(f"\nКоличество пар по интервалам:")
for label, count in corr_dist.items():
    print(f"  {label}: {count} ({count/total_pairs*100:.1f}%)")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew
from sklearn.datasets import load_wine

# Загружаем датасет Wine
wine_data = load_wine()
df = pd.DataFrame(data=wine_data.data, columns=wine_data.feature_names)

print("="*80)
print("ДАТАСЕТ WINE: информация")
print("="*80)
print(f"Размерность данных: {df.shape}")
print(f"Признаки: {', '.join(wine_data.feature_names)}")

# =============================================================================
# ЗАДАНИЕ 2: АНАЛИЗ АСИММЕТРИИ РАСПРЕДЕЛЕНИЙ
# =============================================================================

print("\n" + "="*80)
print("ЗАДАНИЕ 2: АНАЛИЗ АСИММЕТРИИ РАСПРЕДЕЛЕНИЙ (SKEWNESS)")
print("="*80)

# Рассчитываем коэффициент асимметрии для каждого признака
skewness_results = []
for feature in df.columns:
    skew_value = skew(df[feature])
    skewness_results.append({
        'Признак': feature,
        'Коэффициент асимметрии': skew_value,
        'Интерпретация': 'Положительная' if skew_value > 0 else 'Отрицательная' if skew_value < 0 else 'Симметричное'
    })

# Создаем DataFrame с результатами
skewness_df = pd.DataFrame(skewness_results)
skewness_df = skewness_df.sort_values('Коэффициент асимметрии', ascending=False)

print("\nКоэффициенты асимметрии для всех признаков:")
print(skewness_df.to_string(index=False))

# Находим 3 признака с наибольшей положительной асимметрией
top_positive_skew = skewness_df.head(3)

print("\n" + "="*80)
print("ТОП-3 ПРИЗНАКА С НАИБОЛЬШЕЙ ПОЛОЖИТЕЛЬНОЙ АСИММЕТРИЕЙ:")
print("="*80)
print(top_positive_skew[['Признак', 'Коэффициент асимметрии']].to_string(index=False))

# =============================================================================
# ВИЗУАЛИЗАЦИЯ РАСПРЕДЕЛЕНИЙ ТОП-3 ПРИЗНАКОВ
# =============================================================================

# Создаем гистограммы для топ-3 признаков
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Распределения признаков с наибольшей положительной асимметрией', fontsize=16)

for idx, (_, row) in enumerate(top_positive_skew.iterrows()):
    feature = row['Признак']
    skew_value = row['Коэффициент асимметрии']

    ax = axes[idx]

    # Строим гистограмму
    ax.hist(df[feature], bins=20, alpha=0.7, color='steelblue', edgecolor='black', linewidth=0.5)

    # Добавляем вертикальные линии для среднего и медианы
    mean_val = df[feature].mean()
    median_val = df[feature].median()

    ax.axvline(mean_val, color='red', linestyle='-', linewidth=2, label=f'Среднее: {mean_val:.2f}')
    ax.axvline(median_val, color='green', linestyle='--', linewidth=2, label=f'Медиана: {median_val:.2f}')

    ax.set_xlabel(feature, fontsize=11)
    ax.set_ylabel('Частота', fontsize=11)
    ax.set_title(f'{feature}\nКоэффициент асимметрии: {skew_value:.3f}', fontsize=12)
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# =============================================================================
# СРАВНЕНИЕ С ПРИЗНАКАМИ С ОТРИЦАТЕЛЬНОЙ АСИММЕТРИЕЙ
# =============================================================================

print("\n" + "="*80)
print("СРАВНЕНИЕ С ПРИЗНАКАМИ С ОТРИЦАТЕЛЬНОЙ АСИММЕТРИЕЙ:")
print("="*80)

# Находим признаки с отрицательной асимметрией
negative_skew = skewness_df[skewness_df['Коэффициент асимметрии'] < 0].head(3)

if len(negative_skew) > 0:
    print("\nТоп-3 признака с отрицательной асимметрией:")
    print(negative_skew[['Признак', 'Коэффициент асимметрии']].to_string(index=False))

    # Визуализация для сравнения
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle('Распределения признаков с отрицательной асимметрией (для сравнения)', fontsize=16)

    for idx, (_, row) in enumerate(negative_skew.iterrows()):
        feature = row['Признак']
        skew_value = row['Коэффициент асимметрии']

        ax = axes[idx]

        # Строим гистограмму
        ax.hist(df[feature], bins=20, alpha=0.7, color='lightcoral', edgecolor='black', linewidth=0.5)

        # Добавляем вертикальные линии для среднего и медианы
        mean_val = df[feature].mean()
        median_val = df[feature].median()

        ax.axvline(mean_val, color='red', linestyle='-', linewidth=2, label=f'Среднее: {mean_val:.2f}')
        ax.axvline(median_val, color='green', linestyle='--', linewidth=2, label=f'Медиана: {median_val:.2f}')

        ax.set_xlabel(feature, fontsize=11)
        ax.set_ylabel('Частота', fontsize=11)
        ax.set_title(f'{feature}\nКоэффициент асимметрии: {skew_value:.3f}', fontsize=12)
        ax.legend()
        ax.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

# =============================================================================
# ВИЗУАЛИЗАЦИЯ ВСЕХ РАСПРЕДЕЛЕНИЙ (MATRIX PLOT)
# =============================================================================

# Создаем матрицу гистограмм для всех признаков
n_features = len(df.columns)
n_cols = 4
n_rows = (n_features + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, n_rows * 3))
fig.suptitle('Распределения всех признаков датасета Wine', fontsize=16)

# Преобразуем axes в плоский список для удобства
axes_flat = axes.flatten()

for idx, feature in enumerate(df.columns):
    ax = axes_flat[idx]

    # Строим гистограмму
    ax.hist(df[feature], bins=20, alpha=0.7, color='steelblue', edgecolor='black', linewidth=0.5)

    # Получаем коэффициент асимметрии
    skew_value = skewness_df[skewness_df['Признак'] == feature]['Коэффициент асимметрии'].values[0]

    ax.set_xlabel(feature, fontsize=9)
    ax.set_ylabel('Частота', fontsize=9)
    ax.set_title(f'{feature}\nskew = {skew_value:.3f}', fontsize=10)
    ax.grid(True, alpha=0.3)

# Убираем пустые подграфики
for idx in range(len(df.columns), len(axes_flat)):
    axes_flat[idx].axis('off')

plt.tight_layout()
plt.show()

# =============================================================================
# ОБЪЯСНЕНИЕ ВАЖНОСТИ АНАЛИЗА АСИММЕТРИИ
# =============================================================================

print("\n" + "="*80)
print("ПОЧЕМУ ЭТО ВАЖНО ДЛЯ АЛГОРИТМОВ, ЧУВСТВИТЕЛЬНЫХ К РАСПРЕДЕЛЕНИЮ:")
print("="*80)

explanation = """
1. ВЛИЯНИЕ НА МАСШТАБИРОВАНИЕ:
   - Алгоритмы типа SVM с ядром RBF предполагают, что данные имеют приблизительно нормальное распределение
   - Сильная асимметрия может исказить расстояния между точками
   - Признаки с длинным правым хвостом будут иметь выбросы, которые могут сильно влиять на границы решений

2. ПРОБЛЕМЫ С ВЫБРОСАМИ:
   - Положительная асимметрия указывает на наличие выбросов справа
   - Эти выбросы могут стать опорными векторами и исказить разделяющую поверхность
   - SVM пытается максимизировать отступ, и выбросы могут существенно его уменьшить

3. НЕОБХОДИМОСТЬ ПРЕОБРАЗОВАНИЙ:
   - Для признаков с сильной асимметрией рекомендуется применять преобразования:
     * Логарифмическое преобразование (log) - для сильной положительной асимметрии
     * Квадратный корень (sqrt) - для умеренной асимметрии
     * Box-Cox преобразование - автоматический подбор оптимального преобразования
   - Это приближает распределение к нормальному и улучшает работу алгоритмов

4. ИНТЕРПРЕТАЦИЯ РЕЗУЛЬТАТОВ ДЛЯ WINE:
   - В датасете Wine наибольшую положительную асимметрию обычно имеют:
     * 'proline' - содержание пролина (часто имеет длинный правый хвост)
     * 'color_intensity' - интенсивность цвета
     * 'ash' - зольность
   - Эти признаки рекомендуется преобразовывать перед применением SVM
"""

print(explanation)

# =============================================================================
# ДЕМОНСТРАЦИЯ ВЛИЯНИЯ ПРЕОБРАЗОВАНИЯ
# =============================================================================

print("\n" + "="*80)
print("ДЕМОНСТРАЦИЯ ВЛИЯНИЯ ЛОГАРИФМИЧЕСКОГО ПРЕОБРАЗОВАНИЯ:")
print("="*80)

# Берем признак с наибольшей асимметрией
most_skewed_feature = top_positive_skew.iloc[0]['Признак']
original_skew = top_positive_skew.iloc[0]['Коэффициент асимметрии']

# Применяем логарифмическое преобразование (добавляем 1, если есть нули)
if df[most_skewed_feature].min() > 0:
    transformed_data = np.log(df[most_skewed_feature])
    transform_name = 'log'
else:
    # Если есть отрицательные или нулевые значения, добавляем константу
    min_val = df[most_skewed_feature].min()
    transformed_data = np.log(df[most_skewed_feature] - min_val + 1)
    transform_name = 'log(x - min + 1)'

# Рассчитываем новую асимметрию
new_skew = skew(transformed_data)

print(f"\nПризнак: {most_skewed_feature}")
print(f"  - Исходная асимметрия: {original_skew:.3f}")
print(f"  - Асимметрия после {transform_name}: {new_skew:.3f}")
print(f"  - Улучшение: {abs(original_skew) - abs(new_skew):.3f}")

# Визуализация сравнения
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle(f'Сравнение распределений для признака "{most_skewed_feature}"', fontsize=16)

# Исходное распределение
ax1 = axes[0]
ax1.hist(df[most_skewed_feature], bins=20, alpha=0.7, color='steelblue', edgecolor='black')
ax1.axvline(df[most_skewed_feature].mean(), color='red', linestyle='-', linewidth=2, label='Среднее')
ax1.axvline(df[most_skewed_feature].median(), color='green', linestyle='--', linewidth=2, label='Медиана')
ax1.set_xlabel(most_skewed_feature)
ax1.set_ylabel('Частота')
ax1.set_title(f'Исходное распределение\nskew = {original_skew:.3f}')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Преобразованное распределение
ax2 = axes[1]
ax2.hist(transformed_data, bins=20, alpha=0.7, color='lightcoral', edgecolor='black')
ax2.axvline(transformed_data.mean(), color='red', linestyle='-', linewidth=2, label='Среднее')
ax2.axvline(transformed_data.median(), color='green', linestyle='--', linewidth=2, label='Медиана')
ax2.set_xlabel(f'{most_skewed_feature} ({transform_name})')
ax2.set_ylabel('Частота')
ax2.set_title(f'После {transform_name} преобразования\nskew = {new_skew:.3f}')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

# Загружаем датасет Wine
wine_data = load_wine()
df = pd.DataFrame(data=wine_data.data, columns=wine_data.feature_names)
df['target'] = wine_data.target
df['region'] = df['target'].map({0: 'Регион 0', 1: 'Регион 1', 2: 'Регион 2'})

print("="*80)
print("ДАТАСЕТ WINE: информация")
print("="*80)
print(f"Размерность данных: {df.shape}")
print(f"Признаки: {', '.join(wine_data.feature_names)}")
print(f"Регионы: {dict(enumerate(wine_data.target_names))}")

# =============================================================================
# ЗАДАНИЕ 3: СОЗДАНИЕ "ИДЕАЛЬНОГО" РАЗДЕЛЯЮЩЕГО ПРИЗНАКА
# =============================================================================

print("\n" + "="*80)
print("ЗАДАНИЕ 3: СОЗДАНИЕ РАЗДЕЛЯЮЩЕГО ПРИЗНАКА")
print("="*80)

# Создаем новый признак как линейную комбинацию
df['wine_quality_index'] = 0.6 * df['flavanoids'] + 0.4 * df['color_intensity']

print("\nСоздан новый признак: wine_quality_index = 0.6 * flavanoids + 0.4 * color_intensity")
print("\nПервые 10 строк с новым признаком:")
print(df[['region', 'flavanoids', 'color_intensity', 'wine_quality_index']].head(10))

# =============================================================================
# СТАТИСТИКА НОВОГО ПРИЗНАКА ПО РЕГИОНАМ
# =============================================================================

print("\n" + "="*80)
print("СТАТИСТИКА НОВОГО ПРИЗНАКА ПО РЕГИОНАМ:")
print("="*80)

# Группируем по регионам и считаем статистики
region_stats = df.groupby('region')['wine_quality_index'].agg(['count', 'mean', 'std', 'min', 'max'])
print("\n", region_stats.round(3))

# =============================================================================
# ВИЗУАЛИЗАЦИЯ РАСПРЕДЕЛЕНИЯ ПО РЕГИОНАМ
# =============================================================================

# Создаем фигуру с тремя типами визуализации
fig = plt.figure(figsize=(16, 10))
fig.suptitle('Анализ разделяющей способности признака wine_quality_index', fontsize=16)

# 1. Гистограмма распределения по регионам
ax1 = plt.subplot(2, 3, 1)
colors = ['red', 'green', 'blue']
regions = df['region'].unique()

for region, color in zip(regions, colors):
    region_data = df[df['region'] == region]['wine_quality_index']
    ax1.hist(region_data, bins=15, alpha=0.6, color=color, label=region, density=True, edgecolor='black')

ax1.set_xlabel('wine_quality_index')
ax1.set_ylabel('Плотность')
ax1.set_title('Распределение по регионам (гистограмма)')
ax1.legend()
ax1.grid(True, alpha=0.3)

# 2. Boxplot
ax2 = plt.subplot(2, 3, 2)
box_data = [df[df['region'] == region]['wine_quality_index'] for region in regions]
bp = ax2.boxplot(box_data, labels=regions, patch_artist=True)

for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.6)

ax2.set_ylabel('wine_quality_index')
ax2.set_title('Распределение по регионам (boxplot)')
ax2.grid(True, alpha=0.3, axis='y')

# 3. Scatter plot (flavanoids vs color_intensity с цветом по регионам)
ax3 = plt.subplot(2, 3, 3)
for region, color in zip(regions, colors):
    region_data = df[df['region'] == region]
    ax3.scatter(region_data['flavanoids'], region_data['color_intensity'],
               c=color, label=region, alpha=0.7, s=50, edgecolors='black', linewidth=0.5)

# Добавляем линии константного индекса
x_range = np.linspace(df['flavanoids'].min(), df['flavanoids'].max(), 100)
for idx_value in [2, 3, 4, 5]:
    y_values = (idx_value - 0.6 * x_range) / 0.4
    ax3.plot(x_range, y_values, 'k--', alpha=0.3, linewidth=1)
    ax3.text(x_range[-1], y_values[-1], f'index={idx_value}', fontsize=8, alpha=0.5)

ax3.set_xlabel('flavanoids')
ax3.set_ylabel('color_intensity')
ax3.set_title('Исходные признаки с линиями индекса')
ax3.legend()
ax3.grid(True, alpha=0.3)

# 4. Сравнение с исходными признаками (flavanoids)
ax4 = plt.subplot(2, 3, 4)
for region, color in zip(regions, colors):
    region_data = df[df['region'] == region]['flavanoids']
    ax4.hist(region_data, bins=15, alpha=0.6, color=color, label=region, density=True, edgecolor='black')

ax4.set_xlabel('flavanoids')
ax4.set_ylabel('Плотность')
ax4.set_title('Исходный признак: flavanoids')
ax4.legend()
ax4.grid(True, alpha=0.3)

# 5. Сравнение с исходными признаками (color_intensity)
ax5 = plt.subplot(2, 3, 5)
for region, color in zip(regions, colors):
    region_data = df[df['region'] == region]['color_intensity']
    ax5.hist(region_data, bins=15, alpha=0.6, color=color, label=region, density=True, edgecolor='black')

ax5.set_xlabel('color_intensity')
ax5.set_ylabel('Плотность')
ax5.set_title('Исходный признак: color_intensity')
ax5.legend()
ax5.grid(True, alpha=0.3)

# 6. Сводная информация
ax6 = plt.subplot(2, 3, 6)
ax6.axis('off')

# Добавляем сводную таблицу
info_text = "Статистика разделения классов:\n\n"
for region in regions:
    mean_val = df[df['region'] == region]['wine_quality_index'].mean()
    std_val = df[df['region'] == region]['wine_quality_index'].std()
    info_text += f"{region}:\n  mean = {mean_val:.3f}\n  std = {std_val:.3f}\n\n"

ax6.text(0.1, 0.9, info_text, transform=ax6.transAxes, fontsize=10,
         verticalalignment='top', bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))

plt.tight_layout()
plt.show()

# =============================================================================
# КОЛИЧЕСТВЕННАЯ ОЦЕНКА РАЗДЕЛЕНИЯ КЛАССОВ
# =============================================================================

print("\n" + "="*80)
print("КОЛИЧЕСТВЕННАЯ ОЦЕНКА РАЗДЕЛЕНИЯ КЛАССОВ:")
print("="*80)

def calculate_separation_coefficient(data, feature, class1, class2):
    """
    Рассчитывает коэффициент разделения между двумя классами:
    |mean1 - mean2| / (std1 + std2)
    """
    class1_data = data[data['region'] == class1][feature]
    class2_data = data[data['region'] == class2][feature]

    mean1 = class1_data.mean()
    mean2 = class2_data.mean()
    std1 = class1_data.std()
    std2 = class2_data.std()

    separation = abs(mean1 - mean2) / (std1 + std2)
    return separation

# Список признаков для сравнения
features_to_compare = ['flavanoids', 'color_intensity', 'wine_quality_index']
region_pairs = [('Регион 0', 'Регион 1'), ('Регион 0', 'Регион 2'), ('Регион 1', 'Регион 2')]

# Создаем таблицу результатов
results = []

for feature in features_to_compare:
    for region1, region2 in region_pairs:
        sep_coef = calculate_separation_coefficient(df, feature, region1, region2)
        results.append({
            'Признак': feature,
            'Пары регионов': f'{region1} vs {region2}',
            'Коэффициент разделения': sep_coef
        })

results_df = pd.DataFrame(results)

# Создаем сводную таблицу для удобного сравнения
pivot_results = results_df.pivot(index='Признак', columns='Пары регионов', values='Коэффициент разделения')
pivot_results['Средний коэффициент'] = pivot_results.mean(axis=1)
pivot_results = pivot_results.sort_values('Средний коэффициент', ascending=False)

print("\nКоэффициенты разделения для всех пар регионов:")
print(pivot_results.round(3))

# =============================================================================
# ДЕТАЛЬНЫЙ АНАЛИЗ УЛУЧШЕНИЯ
# =============================================================================

print("\n" + "="*80)
print("ДЕТАЛЬНЫЙ АНАЛИЗ УЛУЧШЕНИЯ РАЗДЕЛЕНИЯ:")
print("="*80)

# Сравниваем новый признак с исходными
for feature in ['flavanoids', 'color_intensity']:
    improvement = pivot_results.loc['wine_quality_index', 'Средний коэффициент'] - pivot_results.loc[feature, 'Средний коэффициент']
    percent_improvement = (improvement / pivot_results.loc[feature, 'Средний коэффициент']) * 100
    print(f"\nСравнение с {feature}:")
    print(f"  - Коэффициент разделения (исходный): {pivot_results.loc[feature, 'Средний коэффициент']:.3f}")
    print(f"  - Коэффициент разделения (новый): {pivot_results.loc['wine_quality_index', 'Средний коэффициент']:.3f}")
    print(f"  - Улучшение: {improvement:.3f} ({percent_improvement:.1f}%)")

# Визуализация сравнения
fig, ax = plt.subplots(figsize=(10, 6))

x = np.arange(len(features_to_compare))
width = 0.25
multiplier = 0

for region_pair in region_pairs:
    pair_name = f'{region_pair[0]} vs {region_pair[1]}'
    values = [pivot_results.loc[feature, pair_name] for feature in features_to_compare]
    offset = width * multiplier
    rects = ax.bar(x + offset, values, width, label=pair_name)
    multiplier += 1

ax.set_xlabel('Признаки')
ax.set_ylabel('Коэффициент разделения')
ax.set_title('Сравнение разделяющей способности признаков')
ax.set_xticks(x + width)
ax.set_xticklabels(features_to_compare)
ax.legend(loc='upper right')
ax.grid(True, alpha=0.3, axis='y')

# Добавляем значения на столбцы
for i, (feature, values) in enumerate(zip(features_to_compare, pivot_results[region_pairs[0][0] + ' vs ' + region_pairs[0][1]].values)):
    ax.text(i + width/2, values + 0.02, f'{values:.3f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

# =============================================================================
# ВИЗУАЛИЗАЦИЯ РАЗДЕЛЕНИЯ ДЛЯ ТРУДНОРАЗЛИЧИМЫХ КЛАССОВ
# =============================================================================

# Особое внимание уделяем разделению Регион 1 и Регион 2 (самые похожие)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Сравнение разделения регионов 1 и 2', fontsize=16)

# flavanoids
ax1 = axes[0]
for region, color in zip(['Регион 1', 'Регион 2'], ['green', 'blue']):
    region_data = df[df['region'] == region]['flavanoids']
    ax1.hist(region_data, bins=12, alpha=0.6, color=color, label=region, density=True, edgecolor='black')
ax1.set_xlabel('flavanoids')
ax1.set_ylabel('Плотность')
ax1.set_title(f'flavanoids\nкоэф. = {pivot_results.loc["flavanoids", "Регион 1 vs Регион 2"]:.3f}')
ax1.legend()
ax1.grid(True, alpha=0.3)

# color_intensity
ax2 = axes[1]
for region, color in zip(['Регион 1', 'Регион 2'], ['green', 'blue']):
    region_data = df[df['region'] == region]['color_intensity']
    ax2.hist(region_data, bins=12, alpha=0.6, color=color, label=region, density=True, edgecolor='black')
ax2.set_xlabel('color_intensity')
ax2.set_ylabel('Плотность')
ax2.set_title(f'color_intensity\nкоэф. = {pivot_results.loc["color_intensity", "Регион 1 vs Регион 2"]:.3f}')
ax2.legend()
ax2.grid(True, alpha=0.3)

# wine_quality_index
ax3 = axes[2]
for region, color in zip(['Регион 1', 'Регион 2'], ['green', 'blue']):
    region_data = df[df['region'] == region]['wine_quality_index']
    ax3.hist(region_data, bins=12, alpha=0.6, color=color, label=region, density=True, edgecolor='black')
ax3.set_xlabel('wine_quality_index')
ax3.set_ylabel('Плотность')
ax3.set_title(f'wine_quality_index\nкоэф. = {pivot_results.loc["wine_quality_index", "Регион 1 vs Регион 2"]:.3f}')
ax3.legend()
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# =============================================================================
# ВЫВОДЫ
# =============================================================================

print("\n" + "="*80)
print("ВЫВОДЫ:")
print("="*80)

# Находим лучшее и худшее разделение
best_pair = pivot_results.loc['wine_quality_index'].drop('Средний коэффициент').idxmax()
best_value = pivot_results.loc['wine_quality_index', best_pair]
worst_pair = pivot_results.loc['wine_quality_index'].drop('Средний коэффициент').idxmin()
worst_value = pivot_results.loc['wine_quality_index', worst_pair]

print(f"""
1. РАЗДЕЛЯЮЩАЯ СПОСОБНОСТЬ НОВОГО ПРИЗНАКА:
   - Средний коэффициент разделения: {pivot_results.loc['wine_quality_index', 'Средний коэффициент']:.3f}
   - Лучшее разделение: {best_pair} (коэф. = {best_value:.3f})
   - Худшее разделение: {worst_pair} (коэф. = {worst_value:.3f})

2. СРАВНЕНИЕ С ИСХОДНЫМИ ПРИЗНАКАМИ:
   - flavanoids: средний коэф. = {pivot_results.loc['flavanoids', 'Средний коэффициент']:.3f}
   - color_intensity: средний коэф. = {pivot_results.loc['color_intensity', 'Средний коэффициент']:.3f}
   - wine_quality_index: средний коэф. = {pivot_results.loc['wine_quality_index', 'Средний коэффициент']:.3f}

3. УЛУЧШЕНИЕ РАЗДЕЛЕНИЯ:
""")

for feature in ['flavanoids', 'color_intensity']:
    improvement = pivot_results.loc['wine_quality_index', 'Средний коэффициент'] - pivot_results.loc[feature, 'Средний коэффициент']
    percent = (improvement / pivot_results.loc[feature, 'Средний коэффициент']) * 100
    if improvement > 0:
        print(f"   ✓ Относительно {feature}: +{improvement:.3f} (+{percent:.1f}%)")
    else:
        print(f"   ✗ Относительно {feature}: {improvement:.3f} ({percent:.1f}%)")

print(f"""
4. ИНТЕРПРЕТАЦИЯ:
   - Коэффициент разделения > 0.5 обычно указывает на хорошее разделение
   - Коэффициент разделения > 1.0 указывает на отличное разделение
   - Новый признак {'улучшает' if pivot_results.loc['wine_quality_index', 'Средний коэффициент'] > max(pivot_results.loc['flavanoids', 'Средний коэффициент'], pivot_results.loc['color_intensity', 'Средний коэффициент']) else 'не улучшает'} разделение классов
""")
```
